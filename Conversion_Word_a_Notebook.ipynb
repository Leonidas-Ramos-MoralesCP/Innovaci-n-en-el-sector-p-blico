## Cómo usar este formato en tu curso

Para cada capítulo o material que tengas en Word:

1. Extraer el texto y las imágenes
2. Crear un notebook con alternancia: **Texto → Imagen → Texto → Imagen**
3. Subir el `.ipynb` a GitHub
4. Los estudiantes lo ven directamente en el navegador
